# Signal Generator
We start by building up the signal, without carrier (i.e baseband), containing only the code and no data

```python
# GPS L1 C/A, PRN 1
BITS_PER_CODE = 1023
code = prn_code = ca.code(1, 0, 0, 1, BITS_PER_CODE)
```

```python
# Create the signal
time = 1 # number of ms

signal = np.zeros(time * BITS_PER_CODE)
for i in range(time):
    signal[i*BITS_PER_CODE:i*BITS_PER_CODE+BITS_PER_CODE] = code

# Plot 
timeplot = np.linspace(0, time, BITS_PER_CODE*time)
plt.figure(figsize=[10,6], facecolor='white')
plt.plot(timeplot, signal)
plt.ylabel("Bit value")
plt.ylim([-2, 2])
plt.xlabel("Time [ms]")
plt.title("Signal generated (no carrier)")
plt.tight_layout()
plt.grid()
plt.show()
```

Now we are going to autocorrelate the signal.

```python
autoCorrelation = np.correlate(code, code, mode='full')
#autoCorrelation = autoCorrelation[int(autoCorrelation.size/2):]

# Plot 
#timeplot = np.linspace(0, time, BITS_PER_CODE*time)
plt.figure(figsize=[10,6], facecolor='white')
plt.plot(autoCorrelation)
plt.xlabel("Time [ms]")
plt.title("Signal generated (no carrier)")
plt.tight_layout()
plt.xlim([1019, 1025])
plt.grid()
plt.show()
```
